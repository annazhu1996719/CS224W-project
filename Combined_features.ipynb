```python
from numpy import genfromtxt
import numpy as np
import pandas as pd
```

```python
# Basic Graph features
basic_graph_data = pd.read_csv('datasets/network_properties_combined.csv', escapechar='\\')
# Node2Vec features
note_to_vec_data = pd.read_csv('datasets/network_node2vec.csv', escapechar='\\')
# Central Nodes features
central_nodes = pd.read_csv('datasets/outputs-central-nodes2.csv', escapechar='\\')
# Anonymous walk features
anonymous = pd.read_csv('datasets/outputs-embedding.csv', escapechar='\\')
# Kaggle features
kaggle = pd.read_csv('datasets/chosen_imdb_kaggle.csv', escapechar='\\')
# IMDB rating featues
imdb_rating = pd.read_csv('datasets/graph_imdb_with_rating.csv', escapechar='\\')
# Network metadata(imdb to graph id)
network_metadata = pd.read_csv('datasets/network_metadata.tab.tsv', sep='\t')
# Anonymous walk Probability
anonymous_prob = pd.read_csv('outputs-embedding-awefb.csv', escapechar='\\')
```

```python
# Join all graph properties based on graph_id:
graph_properties = basic_graph_data.set_index('graph_id')\
.join(note_to_vec_data.set_index('graph_id'))\
.join((central_nodes).set_index('graph_id'))\
.join(anonymous.set_index('graph_id'))\
.join(anonymous_prob.set_index('graph_id'))
len(note_to_vec_data), len(basic_graph_data), len(graph_properties)
```

```python
# Join IMDB features with graph property through graph metadata
graph_properties_with_imdb = network_metadata[['GexfID', 'IMDB_id']].set_index('GexfID') \
.join(graph_properties)
graph_properties_with_rating = imdb_rating[['averageRating', 'tconst']].set_index('tconst') \
.join(graph_properties_with_imdb.set_index('IMDB_id'))
graph_properties_with_rating_kaggle_imdb = graph_properties_with_rating \
.join(kaggle.set_index('tid'), how='inner')
len(graph_properties_with_imdb), len(imdb_rating), len(graph_properties_with_rating), \
len(kaggle), len(graph_properties_with_rating_kaggle_imdb)
```

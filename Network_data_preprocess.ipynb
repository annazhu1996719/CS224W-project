```python
from numpy import genfromtxt
import numpy as np
import pandas as pd
```

```python
note_to_vec_data = pd.read_csv('datasets/network_node2vec.csv', escapechar='\\')
```

```python
network_metadata = pd.read_csv('datasets/network_metadata.tab.tsv', sep='\t')
imdb_rating = pd.read_csv('datasets/graph_imdb_with_rating.csv', escapechar='\\')
```

```python
note2vec_with_imdb = network_metadata[['GexfID', 'IMDB_id']].set_index('GexfID').join(note_to_vec_data.set_index('graph_id'))
```

```python
note2vec_with_rating = imdb_rating[['averageRating', 'tconst']].set_index('tconst').join(note2vec_with_imdb.set_index('IMDB_id'))
```

```python
note2vec_with_rating.to_csv("datasets/network_properties_with_rating.csv")
```
